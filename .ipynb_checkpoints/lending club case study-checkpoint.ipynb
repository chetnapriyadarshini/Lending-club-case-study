```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("C:/Users/chetna.priyadarshin/AI_ML_Projects/Credit_worthiness/loan.csv")
df.shape
```

```python
df.head()
```

```python
#Drop null, NaN column
df=df.dropna(axis=1,how='all')
print('After dropping null colns',df.shape)
#Drop below column as it is practically just n, doesn't give us any info
df = df.drop(['pymnt_plan', 'url','desc','initial_list_status','collections_12_mths_ex_med',
              'application_type','policy_code','chargeoff_within_12_mths','tax_liens'],
             axis=1)
print('After dropping colns',df.shape)
```

```python
#Clean the months since delinq column, remove NaN to 0. NaN means that the user is not delinquent. 
#Donwgrade to integer as we are not interested in float values here.

df.loc[df['mths_since_last_delinq'].isnull(),'mths_since_last_delinq'] = 0
df['mths_since_last_delinq'] = df['mths_since_last_delinq'].astype(int)
print(df['mths_since_last_delinq'])
```

```python
#Clean the months since last record column, remove NaN to 0.  
#Donwgrade to integer as we are not interested in float values here.

df.loc[df['mths_since_last_record'].isnull(),'mths_since_last_record'] = 0
df['mths_since_last_record'] = df['mths_since_last_record'].astype(int)
print(df['mths_since_last_record'])
```

```python
#Remove months from colmun value of term and add it to column name
df['term'] = df['term'].str.replace(' months','').astype(int)
df = df.rename(columns={'term': 'Term in Months'})
df.head()
```

```python
#Storing the defaulters in a separate df, might come in handy later on
defaulters_data = df[df['loan_status']=='Charged Off']
defaulters_data.head()
```

```python
#Columns of interest: Verification Status, home_ownership, annual_inc, loan_status='current', term
#Change all float values to int
df['term'] = df['term'].str.replace(' months','').astype(int)
df.to_csv('loan_data_cleaned.csv')
```
